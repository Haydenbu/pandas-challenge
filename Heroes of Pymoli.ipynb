```python
import pandas as pd
import numpy as np
```

```python
pymoli = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(pymoli)
purchase_data.head()
```

```python
player_count = len(purchase_data["SN"].value_counts())
total_players = pd.DataFrame({"Total Players":[player_count]})
total_players
```

```python
unique_items = len(purchase_data["Item ID"].unique())
avg_price = purchase_data["Price"].mean()
num_purchases = purchase_data["Purchase ID"].count()
total_rev = purchase_data["Price"].sum()

sum_table = pd.DataFrame({"Number of Unique Items":[unique_items],
                        "Average Price":avg_price,
                        "Number of Pruchases":num_purchases,
                        "Total Revenue":total_rev
                         })
sum_table
```

```python
gender_count = purchase_data.groupby(["Gender"])
gender_tot = gender_count.nunique()["SN"]
percent_gender = gender_tot/player_count*100
gender_df = pd.DataFrame({"Total Count": gender_tot,
                          "Percentage of Players": percent_gender
                         })
gender_df.index.name = None
gender_df
```

```python
purchase_count = gender_count["Purchase ID"].count()
avg_purch_price = gender_count["Price"].mean()
tot_purch_val = gender_count["Price"].sum()
avg_purch_person = tot_purch_val/gender_tot
purch_df = pd.DataFrame({"Purchase Count":purchase_count,
                        "Average Purchase Price":avg_purch_price,
                        "Total Purchase Value":tot_purch_val,
                        "Avg Total Purchase Per Person":avg_purch_person
                        })
purch_df
```

```python
bins = [0,10,15,20,25,30,35,40,200]
group_names = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Age Group"]=pd.cut(purchase_data["Age"],bins,labels=group_names)
group_age = purchase_data.groupby("Age Group")
count_age_group = group_age["SN"].nunique()
percent_age_group = count_age_group/player_count*100
age_dem_table = pd.DataFrame({"Total Count":count_age_group,
                             "Percentage of Players":percent_age_group})
age_dem_table
```

```python
purch_count_age = group_age["Purchase ID"].count()
avg_purch_price_age = group_age["Price"].mean()
tot_purch_val_age = group_age["Price"].sum()
avg_tot_purch_person_age = tot_purch_val_age/count_age_group
age_sum_table = pd.DataFrame({"Purchase Count":purch_count_age,
                            "Average Purchase Price":avg_purch_price_age,
                            "Total Purchase Value":tot_purch_val_age,
                            "Avg Total Purchase per Person":avg_tot_purch_person_age
                            })
age_sum_table
```

```python
spenders = purchase_data.groupby("SN")
spender_purch_count = spenders["Purchase ID"].count()
spender_avg_price = spenders["Price"].mean()
spender_tot_purch = spenders["Price"].sum()
spender_sum_table = pd.DataFrame({"Purchase Count":spender_purch_count,
                                 "Average Purchase Price":spender_avg_price,
                                 "Total Purchase Value":spender_tot_purch})
sort_spender_sum_table = spender_sum_table.sort_values("Total Purchase Value", ascending=False)
sort_spender_sum_table.head()
```

```python
pop_items = purchase_data[["Item ID", "Item Name", "Price"]]
pop_items_data = pop_items.groupby(["Item ID", "Item Name"])
pop_items_purch_count = pop_items_data["Item ID"].count()
pop_items_price = pop_items_data["Price"].mean()
pop_items_purch_val = pop_items_purch_count*pop_items_price
pop_items_table = pd.DataFrame({"Purchase Count":pop_items_purch_count,
                               "Item Price":pop_items_price,
                               "Total Purchase Value":pop_items_purch_val})
sort_pop_items_table = pop_items_table.sort_values("Purchase Count", ascending=False)
sort_pop_items_table.head()
```
